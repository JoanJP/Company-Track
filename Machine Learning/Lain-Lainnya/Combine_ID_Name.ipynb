# Combine ID & Name

## Import Packages

```python
import pandas as pd
```

## Import Datasets

```python
Branch_df = pd.read_csv('./Dataset/Clean/Branch_clean.csv')
```

```python
Item_df = pd.read_csv('./Dataset/Clean/Item_clean.csv')
```

```python
Order_df = pd.read_csv('./Dataset/Clean/Order_clean.csv')
```

```python
OrderItem_df = pd.read_csv('./Dataset/Clean/OrderItem_clean.csv')
```

## Add Missing Branches & Items

### Merge Datasets

```python
Merged_df = pd.merge(OrderItem_df, Order_df, on='POSNo')
Merged_df
```

### Take Unique IDs

```python
Branches = pd.DataFrame({'BranchID': Merged_df['BranchID'].unique()})
Branches
```

```python
Items = pd.DataFrame({'ItemID': Merged_df['ItemID'].unique()})
Items
```

### Add Missing Values to Datasets

```python
Branch_df = pd.merge(Branch_df, Branches, on='BranchID', how='outer')
Branch_df
```

```python
Item_df = pd.merge(Item_df, Items, on='ItemID', how='outer')
Item_df
```

## Combine ID & Name

### Branch

```python
Branch_df['Branch'] = Branch_df.apply(
    lambda row: row['BranchID']
    if pd.isna(row['BranchName']) 
    else f'{row['BranchID']} - {row['BranchName']}', 
    axis=1)
Branch_df = Branch_df.sort_values(by='BranchID')
Branch_df
```

### Item

```python
Item_df['Item'] = Item_df.apply(
    lambda row: row['ItemID']
    if pd.isna(row['ItemName']) 
    else f'{row['ItemID']} - {row['ItemName']} ({row['BaseUOM']})', 
    axis=1)
Item_df = Item_df.sort_values(by='ItemID')
Item_df
```

## Export Datasets

### Branch

```python
Branch_df.to_csv('Branch_combined.csv', index=False)
```

### Item

```python
Item_df.to_csv('Item_combined.csv', index=False)
```
