# Latest Prices

## Import Packages

```python
import pandas as pd
```

## Import Datasets

```python
Order_df = pd.read_csv('./Dataset/Clean/Order_clean.csv')
Order_df.head()
```

```python
OrderItem_df = pd.read_csv('./Dataset/Clean/OrderItem_clean.csv')
OrderItem_df.head()
```

## Merging Data

```python
Merged_df = pd.merge(OrderItem_df, Order_df, on='POSNo')
Merged_df
```

```python
Daily_df = Merged_df.groupby(['BranchID','ItemID','POSDate']).agg({'Price': 'first'}).sort_values(by=['BranchID','ItemID']).reset_index()
Daily_df = Daily_df.rename(columns={'Price': 'PricePerItem'})
Daily_df = Daily_df.astype({'POSDate': 'datetime64[ns]'})
Daily_df
```

## Get Latest Prices

```python
Prices_df = Daily_df.groupby(['BranchID', 'ItemID']).agg(PricePerItem=('PricePerItem','last')).reset_index()
Prices_df
```

## Export Dataset

```python
Prices_df.to_csv('Prices.csv',index=False)
```
